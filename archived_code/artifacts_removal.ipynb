```python
import os
import numpy as np
from skimage.io import imshow, imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
sample = imread(os.path.join('..', 'data', 'preprocessed_images', '67.png'))
import PIL
from PIL import Image
ctr=0
greyscale=True
img = PIL.Image.open(os.path.join('..', 'data', 'preprocessed_images', '67.png'))
width,height=img.size
for i in range(height):
    for j in range(width):
        a=int(sample[i][j][0])
        b=int(sample[i][j][1])
        c=int(sample[i][j][2])
        if(abs(a-b)>45 or abs(a-c)>45 or abs(b-c)>45):
            print(a,b,c)
            greyscale=False
            listofavg=[]
            #up
            curr=i-1
            while(curr>=0):
                a=int(sample[curr][j][0])
                b=int(sample[curr][j][1])
                c=int(sample[curr][j][2])
                if(not (abs(a-b)>30 or abs(a-c)>30 or abs(b-c)>30)):
                    listofavg.append(sample[curr][j])
                    break
                curr-=1
            #down
            curr=i+1
            while(curr<height):
                a=int(sample[curr][j][0])
                b=int(sample[curr][j][1])
                c=int(sample[curr][j][2])
                if(not (abs(a-b)>30 or abs(a-c)>30 or abs(b-c)>30)):
                    listofavg.append(sample[curr][j])
                    break
                curr+=1
            #left
            curr=j-1
            while(curr>=0):
                a=int(sample[i][curr][0])
                b=int(sample[i][curr][1])
                c=int(sample[i][curr][2])
                if(not (abs(a-b)>30 or abs(a-c)>30 or abs(b-c)>30)):
                    listofavg.append(sample[i][curr])
                    break
                curr-=1
            #right
            curr=j+1
            while(curr<width):
                a=int(sample[i][curr][0])
                b=int(sample[i][curr][1])
                c=int(sample[i][curr][2])
                if(not (abs(a-b)>30 or abs(a-c)>30 or abs(b-c)>30)):
                    listofavg.append(sample[i][curr])
                    break
                curr+=1
            vals=len(listofavg)
            a=0
            b=0
            c=0
            d=0
            for k in listofavg:
                a+=k[0]
                b+=k[1]
                c+=k[2]
                d+=k[3]
            sample[i][j]=[a/vals,b/vals,c/vals,d/vals]
imshow(sample)
print(greyscale)
#IF GREYSCALE IS FALSE, NO NEED RUN THE NEXT PORTION
#IF GREYSCALE IS TRUE, NEED TO RUN THE NEXT PORTION WITH ORIGINAL IMAGE
```
