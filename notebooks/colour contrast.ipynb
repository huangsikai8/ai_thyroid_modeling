```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
#histogram equilization 
```

```python
img = cv.imread(r'A:\1.png')
#print(img.shape)
img1 = cv.imread(r'A:\1.png')
#print(img1.shape)
#img1 = np.transpose(img1, (1279,1925,3)).reshape(data_shape)

images=np.concatenate((img,img1),axis=1)
figure(figsize=(15, 15))
plt.imshow(images)

#cv.waitKey(0)
#cv.destroyAllWindows()
```

```python
#convert to grey scale image
gray_img=cv.cvtColor(images,cv.COLOR_BGR2GRAY)
figure(figsize=(15,15))
plt.imshow(gray_img)
```

```python
#calculate histogram 
hist=cv.calcHist(gray_img,[0],None,[256],[0,256])
plt.title("Image1")
plt.xlabel('bins')
plt.ylabel("No of pixels")
plt.plot(hist)
```

```python
#equalise contrast
gray_img_eqhist=cv.equalizeHist(gray_img)
hist=cv.calcHist(gray_img_eqhist,[0],None,[256],[0,256])
plt.plot(hist)
```

```python
figure(figsize=(15,15))
plt.imshow(gray_img_eqhist)
```

```python
# create a CLAHE object

f, axs = plt.subplots(2,1,figsize=(15,15))
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(gray_img_eqhist)
plt.subplot(211),plt.imshow(cl1),plt.title('contrast limit:2')
clahe = cv.createCLAHE(clipLimit=20.0, tileGridSize=(8,8))
cl2 = clahe.apply(gray_img_eqhist)
plt.subplot(212),plt.imshow(cl2),plt.title('contrast limit:20')

plt.show()
```

```python
f, axs = plt.subplots(2,1,figsize=(15,15))

img_output1 = cv.cvtColor(cl1, cv.COLOR_GRAY2BGR)
plt.subplot(211),plt.imshow(img_output1),plt.title('1')
img_output2 = cv.cvtColor(cl2, cv.COLOR_GRAY2BGR)
plt.subplot(212),plt.imshow(img_output2),plt.title('2')
plt.show
```

```python
#blue green red
image = cv.imread(r'A:\1.png')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
hist=cv.calcHist(image,[0],None,[256],[0,256])
plt.title("Image1")
plt.xlabel('bins')
plt.ylabel("No of pixels")
plt.plot(hist)
```

```python
ycbcr = cv.cvtColor(image, cv.COLOR_BGR2YCR_CB)
plt.imshow(ycbcr)
```

```python


channels=cv.split(ycbcr)
print (len(channels))

#only change brightness
cv.equalizeHist(channels[0],channels[0])
ycbcr00=cv.merge(channels,ycbcr)
one = cv.cvtColor(ycbcr00,cv.COLOR_YCR_CB2BGR)

images=np.concatenate((ycbcr00,one),axis=1)

figure(figsize=(15,15))


plt.imshow(images)

```
