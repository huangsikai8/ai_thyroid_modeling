```python
#averaging: average of pixels in kernal: kernal size bigger, blurrer

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

f, axs = plt.subplots(2,2,figsize=(15,15))

img = cv.imread(r'A:\1.png')
plt.subplot(221),plt.imshow(img),plt.title('Original')

blur = cv.blur(img,(5,5)) #kernal size
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(blur),plt.title('5x5 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.blur(img,(7,7)) #kernal size
plt.subplot(223),plt.imshow(blur),plt.title('7x7 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.blur(img,(11,11)) #kernal size
plt.subplot(224),plt.imshow(blur),plt.title('11x11 kernal')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#gaussian filter - blur based on kernal(must be ) and standard deviation - larger standard deviation more blur
# blurs edges while bluring image

f, axs = plt.subplots(2,2,figsize=(15,15))

img = cv.imread(r'A:\1.png')
plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

blur = cv.GaussianBlur(img,(5,5),0) #kernal size and sigma 
plt.subplot(222),plt.imshow(blur),plt.title('5x5 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.GaussianBlur(img,(11,11),0) #kernal size and sigma 
plt.subplot(223),plt.imshow(blur),plt.title('7x7 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.GaussianBlur(img,(11,11),5) #kernal size and sigma 
plt.subplot(224),plt.imshow(blur),plt.title('11x11 kernal')
plt.xticks([]), plt.yticks([])
plt.show()

```

```python
#median bluring - removes noise well but removes edges also 

f, axs = plt.subplots(2,2,figsize=(15,15))

img = cv.imread(r'A:\1.png')
plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

blur = cv.medianBlur(img,5,5) #kernal size
plt.subplot(222),plt.imshow(blur),plt.title('5x5 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.medianBlur(img,7,7) #kernal size
plt.subplot(223),plt.imshow(blur),plt.title('7x7 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.medianBlur(img,11,11) #kernal size
plt.subplot(224),plt.imshow(blur),plt.title('11x11 kernal')
plt.xticks([]), plt.yticks([])

plt.show()


```

```python
#bilateral filter preserves edges but then does not blur as much 

#minimum aplitude: smaller, edges sharper
#sigma colour: The greater the value, the colors farther to each other will start to get mixed.
#sigma colour The greater its value, the more further pixels will mix together


f, axs = plt.subplots(2,2,figsize=(15,15))

img = cv.imread(r'A:\1.png')
plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

blur = cv.bilateralFilter(img,5,75,75) #minimum amplitude of edge, sigma colour, sigma colour
plt.subplot(222),plt.imshow(blur),plt.title('5x5 kernal')
plt.xticks([]), plt.yticks([])

blur = cv.bilateralFilter(img,7,0,0) #minimum amplitude of edge, sigma colour, sigma colour
plt.subplot(223),plt.imshow(blur),plt.title('7x7 kernal,sigma colour: 0')
plt.xticks([]), plt.yticks([])

blur = cv.bilateralFilter(img,100,0,0) #minimum amplitude of edge, sigma colour, sigma colour
plt.subplot(224),plt.imshow(blur),plt.title('100 kernal,sigma colour: 0')
plt.xticks([]), plt.yticks([])

plt.show()


```
