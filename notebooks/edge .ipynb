```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
img = cv.imread(r'A:\1.png')
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#gray_img=cv.cvtColor(edges,cv.COLOR_BGR2GRAY)
figure(figsize=(15,15))
edges=cv.equalizeHist(edges)
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(edges)
plt.imshow(cl1)
```

```python

img = cv.imread(r'A:\1.png')
f, axs = plt.subplots(3,1,figsize=(15,15))
plt.subplot(411),plt.imshow(img),plt.title('Original')


# Convert to graycsale
#img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Blur the image for better edge detection

img_blur = cv.GaussianBlur(img, (3,3), 0)

# Sobel Edge Detection

sobelx = cv.Sobel(src=img_blur, ddepth=cv.CV_64F, dx=1, dy=0, ksize=3) # Sobel Edge Detection on the X axis
plt.subplot(412),plt.imshow(sobelx),plt.title('sobelx')
sobely = cv.Sobel(src=img_blur, ddepth=cv.CV_64F, dx=0, dy=1, ksize=3) # Sobel Edge Detection on the Y axis
plt.subplot(413),plt.imshow(sobely),plt.title('sobely')
sobelxy = cv.Sobel(src=img_blur, ddepth=cv.CV_64F, dx=1, dy=1, ksize=3) # Combined X and Y Sobel Edge Detection
plt.subplot(414),plt.imshow(sobelxy),plt.title('sobelxy')
# Display Sobel Edge Detection Images

plt.show()
```
